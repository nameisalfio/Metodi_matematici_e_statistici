# Costruire il grafico di una distribuzione di probabilità

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

### Distribuzione binomiale

```python
n, p = 5, 0.4

x = range(0, n+1)
y = binom.pmf(x, n, p)
```

```python
plt.plot(x, y, '--o') 
plt.title(f'Distribuzione Binomiale (n={n}, p={p})')
plt.xlabel("Numero di Successi")
plt.ylabel("Probabilità")
plt.grid()
plt.show()
```

### Funzione di ripartizione della binomiale

```python
x = np.linspace(0, n+1, 100)
y = binom.cdf(x, n, p)

plt.plot(x, y, '--o')
plt.title("Funzione di ripartizione P(X <= x)")
plt.xlabel("numero di successi")
plt.ylabel("funzione di ripartizione")
plt.grid()
plt.show
```
