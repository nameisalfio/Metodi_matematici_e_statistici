# Costruire il grafico di una distribuzione di probabilit√†

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

### Distribuzione binomiale

```python
n, p = 5, 0.4

x = np.arange(0, n+1)
x = range(0, n+1)
y = binom.pmf(x, n, p)
```

```python
plt.plot(x, y, '.') 
plt.title(f'Distribuzione Binomiale (n={n}, p={p})')
plt.xlabel("Numero di Successi")
plt.ylabel("Probabilit√†")
plt.grid()
plt.show()
```

### Funzione di ripartizione della binomiale

```python
x = np.linspace(0, n+1, 100)
y = binom.cdf(x, n, p)

plt.plot(x, y)
plt.title("Funzione di ripartizione P(X <= x)")
plt.xlabel("numero di successi")
plt.ylabel("funzione di ripartizione")
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parametri della distribuzione binomiale
n = 10  # Numero di tentativi
p = 0.3  # Probabilit√† di successo in ciascun tentativo

# Calcola la funzione di ripartizione
k = np.arange(0, n+1)  # Valori possibili della variabile casuale
cdf = binom.cdf(k, n, p)

# Grafico della funzione di ripartizione
plt.plot(k, cdf, marker='o', linestyle='-')
plt.title('Funzione di ripartizione di una distribuzione binomiale')
plt.xlabel('k (Numero di successi)')
plt.ylabel('F(k) (Probabilit√† di k o meno successi)')
plt.grid()
plt.show()
```
