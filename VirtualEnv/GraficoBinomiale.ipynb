# Costruire il grafico di una distribuzione di probabilità

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

### Distribuzione

```python
n, p = 5, 0.4

x = np.arange(0, n+1)
y = binom.pmf(x, n, p)
```

```python
plt.plot(x,y, ".")
plt.title("Densità p(K)")
plt.xlabel("numero di successi")
plt.ylabel("probabilità")
plt.show
```

### Funzione di ripartizione

```python
xx = np.linspace(0, n+2, 100)
z = binom.cdf(xx, n, p)
```

```python
plt.plot(xx,z)
plt.title("Funzione di ripartizione P(z <= xx)")
plt.xlabel("numero di successi")
plt.ylabel("funzione di ripartizione")
plt.show
```
