# Esercizio sulla distribuzione multinomiale

Calcolare la probabilità che lanciando un dado non truccato 4 volte esca 3 volte 6 e 1 volta 2

```python
from scipy.stats import multinomial

n = 4
k = [0, 1, 0, 0, 0, 3]  # Esiti
p = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]  # Probabilità normalizzate per un dado non truccato

P = multinomial.pmf(k, n, p)

P
```

```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.arange(0, n+1)
pmf = multinomial.pmf(k, x, p) 
y = np.cumsum(pmf)
plt.plot(x,y)
plt.show()
```
