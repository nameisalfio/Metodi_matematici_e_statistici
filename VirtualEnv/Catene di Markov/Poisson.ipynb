## Generare numeri pseudocasuali che seguono una distribuzione di Poisson

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def esponenziale(lam, n):
    x = np.random.rand(n)
    y = -np.log(x)/lam
    return y

def my_poisson(lam):
    t, k = 0, 0 #tempo e numero di volte
    while t <= 1:
        r = esponenziale(lam, 1)
        t += r
        k += 1
    k -= 1
    return k

# Genero N numeri che seguono una Poisson
N = 100000
lam = 2
x = np.array([my_poisson(lam) for i in range(N)])

M = np.max(x)
print(f'M: {M}\n')

# Conto quante volte 
F = np.zeros(int(M+1))
for i in range(N):
    k = int(x[i])
    F[k] += 1
F /= N

print(f'F: {F}')

# Istogramma
xx = np.arange(int(M+1))
yy = poisson.pmf(xx, lam)
plt.bar(xx, yy)
plt.plot(xx, F, 'o', color='red')
plt.show()
```
